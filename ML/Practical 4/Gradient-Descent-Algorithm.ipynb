```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function y = (x + 3)^2


def func(x):
    return (x + 3)**2

# Derivative of the function: dy/dx = 2(x + 3)


def derivative(x):
    return 2 * (x + 3)

# Gradient Descent function


def gradient_descent(starting_point, learning_rate, iterations):
    x = starting_point
    history = [x]

    for _ in range(iterations):
        # Calculate the gradient (derivative)
        grad = derivative(x)

        # Update the value of x
        x = x - learning_rate * grad

        # Record the value of x at each step
        history.append(x)

    return x, history


# Initial parameters
starting_point = 2  # Starting from x = 2
learning_rate = 0.1  # Learning rate (step size)
iterations = 50  # Number of iterations

# Run Gradient Descent
minima, history = gradient_descent(starting_point, learning_rate, iterations)

# Output the results
print(f"Local minima found at x = {minima}")
print(f"Function value at minima: y = {func(minima)}")

# Plot the function and the gradient descent steps
x_vals = np.linspace(-10, 4, 100)
y_vals = func(x_vals)

plt.plot(x_vals, y_vals, label='y = (x + 3)^2')
plt.scatter(history, func(np.array(history)), color='red',
            label='Gradient Descent Steps', zorder=5)
plt.title("Gradient Descent to Find Local Minima")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
